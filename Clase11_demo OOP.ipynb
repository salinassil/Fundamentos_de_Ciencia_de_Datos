#### Creando métodos

```python
class Pokemon:
    name = 'Pikachu'
    ot = 'Electric'
    health = 70
    evol=100

    # creamos un método que accede un atributo de clase
    def firePower(self):
        return self.health*self.evol / 2
```

```python
pok1 = Pokemon()
pok1.evol=200
# invocamos el metodo creado
print(pok1.firePower())

# otra forma de invocar el metodo
print(Pokemon.firePower(pok1))
```

```python
pok2 = Pokemon()
# invocamos el metodo creado
print(pok2.firePower())
```

#### Modificadores de accceso

```python
class Pokemon:
    def __init__(self):
        # variable public
        self.name='Pokemon Warrior'
        # variable private
        self.__health=100
        # variable protected
        self._ot='Electric'

    def name(self):
        return self.name

    def health(self):
        return self.__health
```

```python
pok1 = Pokemon()
print(pok1.name)
```

```python
print(pok1.health)
```

```python
print(pok1.ot)
```

```python
class Pokemon:
    # variable privada
    __health = 0

    def __init__(self,health):
        self.__health = health

    # metodo privado
    def __reset_health(self):
        __health = 100
```

```python
pok1 = Pokemon(50)
print(pok1.reset_health())
```

#### Accesadores y mutadores

```python
class Pokemon:

    def __init__(self, health):
        self.__health = health

    # accesador health
    def get_health(self):
        return self.__health

    # mutador health
    def set_health(self, health):
        if 0 <= health <= 100:
            self.__health = health
        else:
            # se podria mejorar con raise Error
            print('Healt debe estar entre 0 y 100')
```

```python
pok1 = Pokemon(150)
# utilizamos accesador
print('Health es', pok1.get_health())

# utilizamos mutador
pok1.set_health(80)
print('Health es', pok1.get_health())
```

```python
# utilizamos la variable directa
print('Healtg es', pok1.health)
```

#### Property Decorator

```python
class Pokemon:
    def __init__(self, health):
        self._health = health

    @property
    def health(self):
        return self._health

    @health.setter
    def health(self, health):
        if 0 <= health <= 100:
            self._health = health
        else:
            # se podria mejorar con raise Error
            print('Healt debe estar entre 0 y 100')
```

```python
pok1 = Pokemon(50)
print('Health es', pok1.health)
# asignamos un valor invalido
pok1.health=120
print('Health es', pok1.health)
# asignamos un valor valido
pok1.health=90
print('Health es', pok1.health)
```

#### Decorador Dataclass

```python
# importamos la libreria dataclass
from dataclasses import dataclass
```

```python
@dataclass
class Pokemon:
    name: str
    health: int
    ot: str

    def sendAttack(self) -> str:
        print('Attacking opponent with 100 MW power')
```

```python
# creando un objeto pokemon
pok1 = Pokemon('Pikachu',100,'Electric')

# imprimiendo su nombre
print(pok1.name)

# enviando un ataque al oponente
pok1.sendAttack()

# representacion string del objeto
print(pok1)
```

### Gestor de empleados con clases


```python
class Empleado:
    def __init__(self, nombre, rol, salario):
        self.nombre = nombre
        self.rol = rol
        self.salario = salario

    def mostrar(self):
        print(f"Empleado: {self.nombre} | Rol: {self.rol} | Salario: ${self.salario:.2f}")

    def actualizar_salario(self, nuevo_monto):
        self.salario = nuevo_monto

    def gana_mas_que(self, promedio):
        return self.salario > promedio
```

```python
# 3. Instanciar 3 empleados y almacenarlos en una lista
empleados = [
    Empleado("Ana García", "Desarrolladora", 3500.0),
    Empleado("Luis Pérez", "Diseñador", 2800.0),
    Empleado("María López", "Gerente", 5000.0)
]
```

```python

# 4. Recorrer la lista y realizar las operaciones
print("--- Información de los Empleados ---")
suma_salarios = 0
for emp in empleados:
    emp.mostrar()
    suma_salarios += emp.salario

# Calcular el promedio de salario
promedio_general = suma_salarios / len(empleados)
print(f"\nEl salario promedio es: ${promedio_general:.2f}")
print("------------------------------------\n")

# Usar gana_mas_que() para informar quién está por encima del promedio
print("Empleados con salario superior al promedio:")
for emp in empleados:
    if emp.gana_mas_que(promedio_general):
        print(f"- {emp.nombre} está por encima del promedio.")
```

```python
class Gerente(Empleado):
    def __init__(self, nombre, rol, salario, departamento):
        super().__init__(nombre, rol, salario)
        self.departamento = departamento

    def mostrar(self):
        super().mostrar()
        print(f"Departamento: {self.departamento}")
```

```python
gerente=Gerente("Ana García", "Desarrolladora", 3500.0, "Ventas")
gerente.mostrar()
```
