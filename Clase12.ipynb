```python

import time

start_time = time.time()  # Marca el tiempo de inicio

lista = [1, 2, 3]
lista1 = []
for i in lista:
   lista1.append(i * 2)

print(lista1)

end_time = time.time()    # Marca el tiempo de finalización

elapsed_time = end_time - start_time
print(f"Tiempo de ejecución: {elapsed_time} segundos")
```
