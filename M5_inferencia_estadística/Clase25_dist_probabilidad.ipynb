```python
import random
import numpy as np
import pandas as pd

n_experimentos = 10000
moneda = random.choices(['cara', 'cruz'], k=n_experimentos)
# Convertimos la lista a una Serie de pandas para usar value_counts()
var=pd.Series(moneda).value_counts()
print(var)
var.plot(kind='bar', title='Frecuencia de resultados', ylabel='Cantidad', xlabel='Resultado', color=['blue', 'orange'])
```
