###Ejemplo merge(), concat() y join()

```python
import pandas as pd

# 1. Dataset de Empleados (La base)
df_empleados = pd.DataFrame({
    'id_empleado': [101, 102, 103, 104],
    'nombre': ['Ana', 'Beto', 'Carla', 'David'],
    'departamento': ['Ventas', 'IT', 'IT', 'Marketing']
})

# 2. Dataset de Salarios (Para probar merge())
# Nota: El id 105 no está en empleados, y el 104 no tiene salario aquí.
df_salarios = pd.DataFrame({
    'id_empleado': [101, 102, 103, 105],
    'nombre': ['Ana', 'Beto', 'Carla', 'David'],
    'salario': [50000, 60000, 55000, 45000]
})

# 3. Dataset de Nuevos Ingresos (Para probar concat())
df_nuevos = pd.DataFrame({
    'id_empleado': [106, 107],
    'nombre': ['Elena', 'Federico'],
    'departamento': ['RRHH', 'Ventas']
})

print("--- DataFrame Empleados ---")
print(df_empleados)
print("\n--- DataFrame Salarios ---")
print(df_salarios)
print("\n--- DataFrame Nuevos Ingresos ---")
print(df_nuevos)
```

```python
nuevo=df_empleados.merge(df_salarios,on=['id_empleado','nombre'],how='outer')
nuevo
```

```python
nuevo1=df_empleados.merge(df_nuevos,on=['id_empleado','nombre'],how='outer')
nuevo1
```

```python
df_total=pd.concat([df_empleados,df_nuevos])
df_total
```

```python
df_total_t=df_total.merge(df_salarios,on=['id_empleado','nombre'],how='outer')
df_total_t
```

```python
df_empleados_idx=df_empleados.set_index('id_empleado')
df_empleados_idx
```

```python
df_bonos=pd.DataFrame({'bono':[1000,2000]},index=[101,102])
df_bonos
```

```python
df_salarios1 = pd.DataFrame({
    'id_empleado': [101, 102, 103, 105],
    'salario': [50000, 60000, 55000, 45000]
})
df_salarios_idx=df_salarios1.set_index('id_empleado')
df_salarios_idx
```

```python
final=df_empleados_idx.join([df_salarios_idx,df_bonos])
final
```
